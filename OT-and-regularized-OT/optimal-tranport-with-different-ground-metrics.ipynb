```python
import numpy as np
import matplotlib.pylab as pl
import ot
import ot.plot
```

## Dataset 1: uniform sampling

```python
n = 20
xs = np.zeros((n, 2))
xs[:, 0] = np.arange(n) + 1
xs[:, 1] = (np.arange(n) + 1) * -0.001 # to make it stricly convex

xt = np.zeros((n, 2))
xt[:, 1] = np.arange(n) + 1

a, b = ot.unif(n), ot.unif(n)

# loss matix
M1 = ot.dist(xs, xt, metric="euclidean")
M1 /= M1.max()

M2 = ot.dist(xs, xt, metric="sqeuclidean")
M2 /= M2.max()

Mp = ot.dist(xs, xt, metric="cityblock")
Mp /= Mp.max()

# data
pl.figure(1, figsize=(7, 3))
pl.clf()
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("Source and target distributions")

# cost matrix
pl.figure(2, figsize=(7, 3))

pl.subplot(1, 3, 1)
pl.imshow(M1, interpolation="nearest")
pl.title("Euclidean cost")

pl.subplot(1, 3, 2)
pl.imshow(M2, interpolation="nearest")
pl.title("Squared Euclidean cost")

pl.subplot(1, 3, 3)
pl.imshow(Mp, interpolation="nearest")
pl.title("L1 cityblock cost")
pl.tight_layout()
```

## Dataset 1: Plot OT Matrices

```python
G1 = ot.emd(a, b, M1)
G2 = ot.emd(a, b, M2)
Gp = ot.emd(a, b, Mp)

# OT matrices
pl.figure(3, figsize=(7, 3))

pl.subplot(1, 3, 1)
ot.plot.plot2D_samples_mat(xs, xt, G1, c=[0.5, 0.5, 1])
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("OT Euclidean")

pl.subplot(1, 3, 2)
ot.plot.plot2D_samples_mat(xs, xt, G2, c=[0.5, 0.5, 1])
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("OT squared Euclidean")

pl.subplot(1, 3, 3)
ot.plot.plot2D_samples_mat(xs, xt, Gp, c=[0.5, 0.5, 1])
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("OT L1 cityblock")
pl.tight_layout()

pl.show()
```

## Dataset 2: Partial circle

```python
n = 20
xtot = np.zeros((n + 1, 2))
xtot[:, 0] = np.cos((np.arange(n + 1) + 1.0) * 0.8 / (n + 2) * 2 * np.pi)
xtot[:, 1] = np.sin((np.arange(n + 1) + 1.0) * 0.8 / (n + 2) * 2 * np.pi)

xs = xtot[:n, :]
xt = xtot[1:, :]

a, b = ot.unif(n), ot.unif(n)

# loss matrix
M1 = ot.dist(xs, xt, metric="euclidean")
M1 /= M1.max()

M2 = ot.dist(xs, xt, metric="sqeuclidean")
M2 /= M2.max()

Mp = ot.dist(xs, xt, metric="cityblock")
Mp /= Mp.max()

# data
pl.figure(4, figsize=(7, 3))
pl.clf()
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("Source and target distributions")

# cost matrix
pl.figure(5, figsize=(7, 3))

pl.subplot(1, 3, 1)
pl.imshow(M1, interpolation="nearest")
pl.title("Euclidean cost")

pl.subplot(1, 3, 2)
pl.imshow(M2, interpolation="nearest")
pl.title("Squared Euclidean cost")

pl.subplot(1, 3, 3)
pl.imshow(Mp, interpolation="nearest")
pl.title("L1 cityblock cost")
pl.tight_layout()
```

```python
G1 = ot.emd(a, b, M1)
G2 = ot.emd(a, b, M2)
Gp = ot.emd(a, b, Mp)

# OT matrices
pl.figure(6, figsize=(7, 3))

pl.subplot(1, 3, 1)
ot.plot.plot2D_samples_mat(xs, xt, G1, c=[0.5, 0.5, 1])
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("OT Euclidean")

pl.subplot(1, 3, 2)
ot.plot.plot2D_samples_mat(xs, xt, G2, c=[0.5, 0.5, 1])
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("OT squared Euclidean")

pl.subplot(1, 3, 3)
ot.plot.plot2D_samples_mat(xs, xt, Gp, c=[0.5, 0.5, 1])
pl.plot(xs[:, 0], xs[:, 1], "+b", label="Source sample")
pl.plot(xt[:, 0], xt[:, 1], "xr", label="Target sample")
pl.axis("equal")
pl.title("OT L1 cityblock")
pl.tight_layout()

pl.show()
```
